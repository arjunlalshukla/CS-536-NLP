```python
import pandas as pd
data = pd.read_csv('iris.csv',header=None, names=['sepal length', 'sepal width', 'petal length', 'petal width', 'class'])
_data=data.drop(columns=['class'])

_data.describe(include='all')
```

```python
from sklearn import preprocessing
data_scaled = preprocessing.scale(_data)
data_scaled = pd.DataFrame(data_scaled,columns=['sepal length', 'sepal width', 'petal length', 'petal width'])
data_scaled.head()
data_scaled.describe(include='all')
```

```python
#covariance matrix
data_scaled.cov()
```

```python
#correlation matrix
data.corr()
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=4).fit(data_scaled)
print('PC_components=', pca.components_)
print('variance_ratio=', pca.explained_variance_ratio_)
print('variance=', pca.explained_variance_)
```

```python
pcadf = pd.DataFrame(pca.transform(data_scaled), columns=['PC0', 'PC1', 'PC2', 'PC3'], index=data_scaled.index)
data_scaled['PC0'] = pcadf['PC0']
data_scaled['PC1'] = pcadf['PC1']
data_scaled['PC2'] = pcadf['PC2']
data_scaled['PC3'] = pcadf['PC3']
```

You can assume the dependent features and the independent feature binary. Please implements the Gini index and the entropy gain ratio. You can read the code available online, but you cannot plagiarize it.

```python
from collections import Counter
from math import log2
class Node:
    def __init__(self, data, features, select_func):
        self.data          = data
        self.select_func   = select_func
        self.coll_func     = Node.gain_ratio if select_func == "entropy" else Node.gini_coll
        self.sing_func     = Node.entropy if select_func == "entropy" else Node.gini
        self.cnt           = Counter([x[-1] for x in data])
        self.features      = features
        self.split_feature = None
        self.split_values  = [None] * self.features
        self.children      = None
        
    def split(self):
        if self.select_func == "entropy":
            if self.entropy() < .1:
                return
            info_gain = 0
            for i in range(self.features):
                nodes = self.split_on_feature(i)
                gain  = self.coll_func(self, nodes)
                if gain > info_gain:
                    info_gain = gain
                    self.children  = nodes
                    self.split_feature = i
            if info_gain == 0:
                self.children = None
                return
        else:
            if self.gini() < .05:
                return
            coll_imp = float('inf')
            for i in range(self.features):
                nodes = self.split_on_feature(i)
                imp   = self.coll_func(self, nodes)
                if imp < coll_imp:
                    coll_imp = imp
                    self.children  = nodes
                    self.split_feature = i
        for child in self.children:
            child.split()
        
    def split_on_feature(self, feature):
        self.data = sorted(self.data, key=lambda x: x[feature])
        min_imp  = float('inf')
        retval = None
        for i in range(len(self.data)-1):
            left  = Node(self.data[:i+1], self.features, self.select_func)
            right = Node(self.data[i+1:], self.features, self.select_func)
            imp  = (len(left.data)*self.sing_func(left) + len(right.data)*self.sing_func(right)) / len(self.data)
            if imp < min_imp:
                min_imp  = imp
                self.split_values[feature] = (self.data[i][feature] + self.data[i+1][feature])/2
                retval = (left, right)
        return retval
    
    def rf(self, _class):
        return self.cnt[_class]/len(self.data)
    
    def gini_coll(self, nodes):
        return sum([len(node.data)*node.gini() for node in nodes]) / len(self.data)
        
    def gini(self):
        return 1-sum([self.rf(_class)**2 for _class in self.cnt.keys()])
    
    def gain_ratio(self, nodes):
        diff = self.entropy() - sum([node.entropy() for node in nodes])
        norm =  sum([len(node.data)/len(self.data)*log2(len(node.data)/len(self.data)) for node in nodes])
        return diff / -norm
    
    def entropy(self):
        return -sum([self.rf(_class)*log2(self.rf(_class)) for _class in self.cnt.keys()])
    
    def classify(self, datum):
        if not self.children:
            return self.cnt.most_common()[0][0]
        elif datum[self.split_feature] <= self.split_values[self.split_feature]:
            return self.children[0].classify(datum)
        else:
            return self.children[1].classify(datum)
    
data  = data.sample(frac=1).reset_index(drop=True)

num = 90

train = data[:num].values
test  = data[num:].values
print(len(train), len(test))
node  = Node(train, 4, "gini")
node.split()

#accuracy
hypo = [node.classify(datum[:-1]) for datum in test]
foo = [print((x, y)) for (x, y) in [(hypo[i], test[i]) for i in range(len(test))] if x != y[-1]]
print((len(data) - num - len(foo))/(len(data) - num))

train = data[:num].values
test  = data[num:].values
print(len(train), len(test))
node  = Node(train, 4, "entropy")
node.split()

#accuracy
hypo = [node.classify(datum[:-1]) for datum in test]
foo = [print((x, y)) for (x, y) in [(hypo[i], test[i]) for i in range(len(test))] if x != y[-1]]
print((len(data) - num - len(foo))/(len(data) - num))
```
